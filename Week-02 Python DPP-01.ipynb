Question 3] Write a program to check if a number is even or odd.
Test Input: Number = 12
Expected Output: 12 is an even number.

```python
#Ans]

b = int(input("Enter a number:"))
if b % 2 == 0:
    print(f"{b} is an even number.")
else:
    print(f"{b} is an odd number.")
```

Question 5] Write a Python program using a for loop to calculate the sum of numbers from 1 to 20.
Expected Output: Sum: 210

```python
#Ans]

sum = 0

for i in range(1, 21):
    sum += i

print("Sum:", sum)
```

Question 8] Write a Python program to print all even numbers from this list using a for loop and if statement:
List = [2, 5, 8, 11, 14, 17, 20]

```python
#Ans]

my_list = [2, 5, 8, 11, 14, 17, 20]
for i in my_list:
    if i % 2 == 0:
        print(i)
```

Question 9] Write a program to find the largest number among three numbers taken from the user using if-elif-else
statements.
Test Input: Numbers = 15, 42, 27
Expected Output: The largest number is 42

```python
#Ans] 

a, b, c = 15, 42, 27

if a >= b and a >= c:
    largest = a
elif b >= a and b >= c:
    largest = b
else:
    largest = c

print("The largest number is", largest)
```
