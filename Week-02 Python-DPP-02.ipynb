QUESTION 4] Write a Python program to use a lambda function to double all numbers in the list .

```python
#Ans] 

mylist = [2, 4, 6, 8]

# Using map() with a lambda function to double each number
double = list(map(lambda x: x * 2, mylist))
print(f"Original list: {mylist}")
print(f"Doubled list: {double}")
```

QUESTION 5] Write a program that asks the user to divide two numbers. Handle the case when the user tries to divide by
zero using try-except.
Test Input: Number 1 = 10, Number 2 = 0

```python
# Ans] 

try:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    
    result = num1 / num2
    print(f"The result of {num1} divided by {num2} is: {result}")

except ZeroDivisionError:
    print("Error: Cannot divide by zero. Please enter a non-zero second number.")
```

QUESTION 7] Write a Python program to use a lambda function with filter() to extract even numbers from the list [3,6,9,12,15,18]

```python
numbers = [3, 6, 9, 12, 15, 18]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print(f"Original list: {numbers}")
print(f"Even numbers: {even_numbers}")
```

QUESTION 8] Write a program to take two inputs from the user and divide them. Handle both ValueError (if input is not a
number) and ZeroDivisionError.
Test Input: Number 1 = “abc”, Number 2 = 0

```python
try:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    
    result = num1 / num2
    print(f"The result of {num1} divided by {num2} is: {result}")

except ZeroDivisionError:
    print("Cannot divide by zero.")
    
except ValueError:
    print("Error: Invalid input. Please enter numbers only.")
```
